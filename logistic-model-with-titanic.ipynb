# Data Wrangling

* Data Cleaning
* Encoding

```python
#Checking NULL

df_train.isnull().sum()
```

```python
sns.heatmap(df_train.isnull(),yticklabels = False)
```

```python
df_train.drop('Cabin', axis=1, inplace = True)
```

```python
sns.boxplot(x= 'Pclass', y = 'Age', data = df_train)
```

We can either impute this with median values or remove na, for this we'll remove NA

```python
#Removing null values

df_train.dropna(inplace =True)
```

```python
df_train.isnull().sum()
```

**Data Preprocessing**

```python
#One Hot Encoding for Gender (as it has only 2 classes)

sex = pd.get_dummies(df_train['Sex'],drop_first=True)
sex.head()
```

```python
#One Hot Encoding for Embark (as it has only 3 classes)

embarked = pd.get_dummies(df_train['Embarked'],drop_first=True)
embarked.head()
```

```python
#One Hot Encoding for Pclass (as it has only 2 classes)

pclass = pd.get_dummies(df_train['Pclass'],drop_first=True)
pclass.head()
```

```python
#Concatenating the new encoded features

df_train = pd.concat([df_train,sex,embarked,pclass],axis=1)
```

```python
df_train.head()
```

```python
#dropping the irrelevant col

df_train.drop(['PassengerId','Pclass','Name','Ticket','Embarked'],axis=1,inplace=True)
```

```python
df_train.drop('Sex',axis=1,inplace= True)
```

```python
df_train.head()
```

# Model Building


```python
#Train Test

y = df_train['Survived']
X = df_train.drop('Survived',axis =1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,y,test_size = 0.2,random_state =1)
```

```python
#Model Building

from sklearn.linear_model import LogisticRegression
```

```python
#Defining the model
model = LogisticRegression()
```

```python
#Fitting data into model and training the model
model.fit(X_train, y_train)
```

```python
#Using the model to predict from test data
predictions = model.predict(X_test)
```

# Model Validation

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
